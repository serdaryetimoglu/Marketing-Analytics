```r
boxplot(EuropeanSales1$Population , ylab = 'Population' , 
        main = 'Distiribution Of Population')

boxplot(EuropeanSales1$GDPperHead , ylab = 'GDPperHead' , 
        main = 'Distiribution Of GDPperHead')

boxplot(EuropeanSales1$GDPperHead , ylab = 'UnemploymentRate' , 
        main = 'Distiribution Of UnemploymentRate')

boxplot(EuropeanSales1$GDPperHead , ylab = 'EducationSpending' , 
        main = 'Distiribution Of EducationSpending')

boxplot(EuropeanSales1$GDPperHead , ylab = 'SalesPerCapita' , 
        main = 'Distiribution Of SalesPerCapita')

boxplot(EuropeanSales1$GDPperHead , ylab = 'ComputerSales' , 
        main = 'Distiribution Of ComputerSales')
```

and let see corr matrix

```r
install.packages("corrplot")
library(corrplot)
corrs = cor(EuropeanSales1, method = "pearson")
corrplot(corrs, type = "full", method="ellipse", tl.col="black", 
         tl.srt = 30, addCoef.col = "black")
```

```r
#Models for SalesPerCapita
model1 <- lm(EuropeanSales1$SalesPerCapita ~ EuropeanSales1$GDPperHead + EuropeanSales1$EducationSpending, 
             data = EuropeanSales1)
summary(model1)

model2 <- lm(EuropeanSales1$SalesPerCapita ~ EuropeanSales1$GDPperHead + EuropeanSales1$EducationSpending +
             EuropeanSales1$Population ,
             data = EuropeanSales1)
summary(model2)

model3 <- lm(EuropeanSales1$SalesPerCapita ~ EuropeanSales1$GDPperHead + EuropeanSales1$EducationSpending +
             EuropeanSales1$Population + EuropeanSales1$UnemploymentRate ,
             data = EuropeanSales1)
summary(model3)
```

```r
#Models for ComputerSales
model4 <- lm(EuropeanSales1$ComputerSales ~ EuropeanSales1$Population + EuropeanSales1$UnemploymentRate, 
             data = EuropeanSales1)
summary(model4)

model5 <- lm(EuropeanSales1$ComputerSales ~ EuropeanSales1$Population + EuropeanSales1$UnemploymentRate +
             EuropeanSales1$GDPperHead ,
             data = EuropeanSales1)
summary(model5)

model6 <- lm(EuropeanSales1$ComputerSales ~ EuropeanSales1$Population + EuropeanSales1$UnemploymentRate +
             EuropeanSales1$GDPperHead + EuropeanSales1$EducationSpending ,
             data = EuropeanSales1)
summary(model6)
```
